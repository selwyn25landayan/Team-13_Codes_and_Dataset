```python
import cv2
import mediapipe as mp
import time

# Initialize MediaPipe pose solution
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()

# Open video file
cap = cv2.VideoCapture(r'C:\Users\Renzo\Documents\Final Design\Datasets\train\bicep_curl\bicep curl_3.mp4')

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

# Get video details
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(r'C:\Users\Renzo\Documents\Final Design\output.mp4', fourcc, fps, (width, height))

pTime = 0

while True:
    success, img = cap.read()
    if not success:
        print("Error: Could not read frame.")
        break

    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = pose.process(imgRGB)

    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
        for id, lm in enumerate(results.pose_landmarks.landmark):
            h, w, c = img.shape
            cx, cy = int(lm.x * w), int(lm.y * h)
            cv2.circle(img, (cx, cy), 5, (255, 0, 0), cv2.FILLED)

    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime

    cv2.putText(img, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN, 3, (255, 0, 0), 3)

    # Write the frame into the file 'output.mp4'
    out.write(img)

    cv2.imshow("Image", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import os

# Initialize MediaPipe pose solution
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()

# Folder containing input images
input_folder = r'C:\Users\Renzo\Documents\Final Design\Testing\case 1.4 - 3.4 overhead press'
output_folder = r'C:\Users\Renzo\Documents\Final Design\Testing\overhead'

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Get list of all image files in the input folder
image_files = [f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))]

for image_file in image_files:
    # Load image frame
    img_path = os.path.join(input_folder, image_file)
    img = cv2.imread(img_path)

    if img is None:
        print(f"Error: Could not load image frame {img_path}.")
        continue

    # Convert the image to RGB
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = pose.process(imgRGB)

    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
        for id, lm in enumerate(results.pose_landmarks.landmark):
            h, w, c = img.shape
            cx, cy = int(lm.x * w), int(lm.y * h)
            cv2.circle(img, (cx, cy), 5, (255, 0, 0), cv2.FILLED)

    # Save the processed frame as an image
    output_path = os.path.join(output_folder, f'processed_{image_file}')
    cv2.imwrite(output_path, img)

    print(f"Processed frame saved at: {output_path}")
```

```python
import cv2
import mediapipe as mp

#Initialize MediaPipe pose solution
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()

#Load image file
image_path = r"C:\Users\Renzo\Desktop\screenshots\case 1.1 deadlift\Trial 1.png"
img = cv2.imread(image_path)

if img is None:
    print("Error: Could not open image file.")
    exit()

#Convert the BGR image to RGB
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
results = pose.process(imgRGB)

#Check if any landmarks were found
if results.pose_landmarks:
    mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
    h, w, c = img.shape
    for id, lm in enumerate(results.pose_landmarks.landmark):
        cx, cy = int(lm.x * w), int(lm.y * h)
        print(f'keypoint {id}: x={lm.x:.4f}, y={lm.y:.4f}, z={lm.z:.4f}')
        cv2.circle(img, (cx, cy), 5, (255, 0, 0), cv2.FILLED)
        if id == 32:  # After printing the last landmark (index 32)
            break

#Display the image with keypoints
cv2.imshow("Image", img)
cv2.destroyAllWindows()
```
